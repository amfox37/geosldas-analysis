```python
import numpy
import xarray as xr
import matplotlib.pyplot as plt
 
# Read binary files and reshape to correct size
# The number of rows and columns are in the file name
lats = numpy.fromfile('../test_data/EASE2_M36km.lats.964x406x1.double', 
                      dtype=numpy.float64).reshape((406,964))
lons = numpy.fromfile('../test_data/EASE2_M36km.lons.964x406x1.double', 
                      dtype=numpy.float64).reshape((406,964))
 
# Extract latitude and longitude
# for a given row and column 
grid_row = 46
grid_column = 470
 
lat_val = lats[grid_row, grid_column]
lon_val = lons[grid_row, grid_column]

print('Latitude:', lat_val)
print('Longitude:', lon_val)


print('Lats',lats[:,1])
print('Lons',lons[1,:])
```

```python

ds = xr.open_dataset('DAv7_M36.inst3_1d_lndfcstana_Nt.20150901.nc4')
lon = ds['lon']
lat = ds['lat']
n_tile = len(lat)

# Convert to numpy array
lon = lon.values
lat = lat.values

# Make scatter plot of lon vs lat
plt.figure(figsize=(8,8))
plt.scatter(lon, lat, s=0.1)
```

```python
print('Lats',lats[1:100,1])
print('Lons',lons[1,1:10])

print('Lat',lat[1:10])
print('Lon',lon[1:100])
```

```python
# Make an empty array with dimensions of the grid
grid = numpy.full((406, 964), numpy.nan)
lats_row = lats[:,1]
lons_col = lons[1,:]

# Fill the grid with the values from the dataset
for i in range(n_tile):
    # Find the row and column of the grid
    row = numpy.abs(lats_row - lat[i]).argmin()
    col = numpy.abs(lons_col - lon[i]).argmin()

    # Check if row and col are within the valid range of indices for the grid array
    if row < grid.shape[0] and col < grid.shape[1]:
        grid[row, col] = 1
    else:
        print('Row or column index out of bounds')

# Plot the grid
plt.imshow(grid, cmap='gray', origin='lower', extent=[lons.min(), lons.max(), lats.min(), lats.max()])

```
