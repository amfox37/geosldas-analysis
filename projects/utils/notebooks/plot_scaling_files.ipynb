```python
import numpy as np

with open('./OLv7_M36_MULTI_L1C_zscore_stats_A_p40.txt', 'r') as f:
# with open('./L4SM_OLv7M36_L1C_zscore_stats_A_p40.txt', 'r') as f:    
    lines = f.readlines()

# Separate lines containing strings and numbers
string_lines = [line for line in lines if any(c.isalpha() for c in line)]
number_lines = [line for line in lines if not any(c.isalpha() for c in line)]

# Store the variable names
variable_names = [line.strip() for line in string_lines]

# Convert the number lines to a numpy array
data = np.array([list(map(float, line.split())) for line in number_lines])
```

```python
# Print the variable names, and the max and min values for each variable
for i, variable in enumerate(variable_names):
    print(variable)
    print(f'Max: {data[i, :].max()}')
    print(f'Min: {data[i, :].min()}')
    print()
```

```python
# Make a dictionary with the variable names as keys and rows as values
data_dict = {variable: row for variable, row in zip(variable_names, data)}
```

```python
# Use plot_global_tight to plot mean_obs
from mapper_functions import plot_global_tight

lons = data_dict['sclprm_lon']
lats = data_dict['sclprm_lat']
mean_obs = data_dict['mean_obs']
mean_mod = data_dict['mean_mod']
tile_id = data_dict['sclprm_tile_id']

# Sort by tile_id
sort_idx = np.argsort(tile_id)
lons = lons[sort_idx]
lats = lats[sort_idx]
mean_obs = mean_obs[sort_idx]
mean_mod = mean_mod[sort_idx]
tile_id = tile_id[sort_idx]

# Replace missing values with np.nan
mean_obs[mean_obs == -9999] = np.nan
mean_mod[mean_mod == -9999] = np.nan

print('Mean of mean_obs:', np.nanmean(mean_obs))
print('Std Dev of mean_obs:', np.nanstd(mean_obs))
print('Mean of mean_mod:', np.nanmean(mean_mod))
print('Std Dev of mean_mod:', np.nanstd(mean_mod))

obarray = np.empty([len(lons), 3])
obarray.fill(np.nan)
obarray[:, 1] = lons
obarray[:, 2] = lats
obarray[:, 0] = mean_obs

plot_global_tight(obarray, False, 'OLv7_M36_MULTI_L1C_zscore_stats_A_p40: Mean Obs', 'K', 125, 300)

obarray[:, 0] = mean_mod

plot_global_tight(obarray, False, 'OLv7_M36_MULTI_L1C_zscore_stats_A_p40: Mean Mod', 'K', 125, 300)
```

```python
# Make histograms of the mean_obs and mean_mod values
import matplotlib.pyplot as plt

plt.hist(mean_obs[~np.isnan(mean_obs)], bins=100, alpha=0.5, label='mean_obs')
plt.hist(mean_mod[~np.isnan(mean_mod)], bins=100, alpha=0.5, label='mean_mod')
plt.legend()
plt.title('OLv7_M36_MULTI_L1C_zscore_stats_A_p40')
plt.show()
```

```python
# Using xarray to open and read the variables, dimensions etc from the file ../test_data/OLv7_M36_MULTI_ASCAT_zscore_stats_all_pentads.nc4

import xarray as xr

ds = xr.open_dataset('../test_data/OLv7_M36_MULTI_ASCAT_zscore_stats_all_pentads.nc4')
print(ds)

# Extract the lower left corner of the grid cell
ll_lon = ds['ll_lon']
ll_lat = ds['ll_lat']
d_lon = ds['d_lon']
d_lat = ds['d_lat']

# Extract the mean_obs and mean_mod variables for the 40th pentad
mean_obs3 = ds['o_mean'][39, :, :]
mean_mod3 = ds['m_mean'][39, :, :]

lons = np.arange(ll_lon, ll_lon + 1440 * d_lon, d_lon)
lats = np.arange(ll_lat, ll_lat + 720 * d_lat, d_lat)

# Convert the 2D arrays to 1D arrays
lons = np.repeat(lons, 720)
lats = np.tile(lats, 1440)
mean_obs3 = mean_obs3.values.flatten()
mean_mod3 = mean_mod3.values.flatten()


```

```python

obarray = np.empty([len(lons), 3])
obarray.fill(np.nan)
obarray[:, 1] = lons
obarray[:, 2] = lats

obarray[:, 0] = mean_obs3
plot_global_tight(obarray, False, 'OLv7_M36_MULTI_ASCAT_zscore_stats_all_pentads: P40 Mean Obs', 'SFDS', 0, 1)

obarray[:, 0] = mean_mod3
plot_global_tight(obarray, False, 'OLv7_M36_MULTI_ASCAT_zscore_stats_all_pentads: P40 Mean Mod', 'm3 m3', 0, 1)

```
