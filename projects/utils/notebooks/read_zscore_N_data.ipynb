```python
# Open the netcdf file using xarray
import xarray as xr
import numpy as np

# Open the netcdf file
ds = xr.open_dataset('/Users/amfox/Desktop/GEOSldas_diagnostics/test_data/M36_zscore_stats_2015_doy91_2021_doy90_W_75d_Nmin_5_sp_ALL_all_pentads.nc4')

# Print the headers/variable names of the file
print(ds)

# Extract the variable called n_data
n_data = ds['n_data']
```

```python

# Put this into a 3D numpy array
n_data_array = n_data.values

# Print the shape of the array
print("Size of array: ", n_data_array.shape)

# Print the mean and standard deviation of the array excluding NaN values
print(f"Mean: {np.nanmean(n_data_array):.2f}")
print(f"StdDev: {np.nanstd(n_data_array):.2f}")

# Print how many none NaN values are in the array
print("Number of none NaN values: ", np.sum(~np.isnan(n_data_array)))

# Print how many values in the array are < 6
print("Number of values less than 6: ", np.sum(n_data_array < 6))
print("Number of values less than 11", np.sum(n_data_array < 11))
print("Number of values less than 21", np.sum(n_data_array < 21))

# Print the fraction of values in the array that are < 6
print(f"Fraction of values less than 6: {np.sum(n_data_array < 6) / np.sum(~np.isnan(n_data_array)) * 100:.2f}%")
print(f"Fraction of values less than 11: {np.sum(n_data_array < 11) / np.sum(~np.isnan(n_data_array)) * 100:.2f}%")
print(f"Fraction of values less than 21: {np.sum(n_data_array < 20) / np.sum(~np.isnan(n_data_array)) * 100:.2f}%")
```

```python
# Plot a pdf of the data
import matplotlib.pyplot as plt

plt.hist(n_data_array[~np.isnan(n_data_array)], bins=100)
xlabel = plt.xlabel('n_data')
ylabel = plt.ylabel('Frequency')
title = plt.title('Histogram of n_data')
plt.show()
```

```python
# Plot a pdf of the values below 100

plt.hist(n_data_array[n_data_array < 80], bins=20)
xlabel = plt.xlabel('n_data')
ylabel = plt.ylabel('Frequency')
title = plt.title('Histogram of n_data values < 60')
plt.show()
```

```python
# Loop over the pentads and print the mean and standard deviation of the n_data values and number of values < 6

fraction_less_than_6 = np.zeros(73)
fracttion_less_than_11 = np.zeros(73)
fraction_less_than_21 = np.zeros(73)

for i in range(0, 73):
    pentad = n_data_array[i,:,:]
    print("Pentad: ", i)
    print("Mean: ", np.nanmean(pentad))
    print("StdDev: ", np.nanstd(pentad))
    print("Number of values less than 6: ", np.sum(pentad < 6))
    print("Number of values less than 11: ", np.sum(pentad < 11))
    print("Number of values less than 21: ", np.sum(pentad < 21))
    print("Fraction of values less than 6: ", np.sum(pentad < 6)/np.sum(~np.isnan(pentad)) * 100,"%")
    print("Fraction of values less than 11: ", np.sum(pentad < 11)/np.sum(~np.isnan(pentad)) * 100,"%")
    print("Fraction of values less than 21: ", np.sum(pentad < 20)/np.sum(~np.isnan(pentad)) * 100,"%")
    print("Number of none NaN values: ", np.sum(~np.isnan(pentad)))
    print("")

    fraction_less_than_6[i] = np.sum(pentad < 6)/np.sum(~np.isnan(pentad)) * 100
    fracttion_less_than_11[i] = np.sum(pentad < 11)/np.sum(~np.isnan(pentad)) * 100
    fraction_less_than_21[i] = np.sum(pentad < 21)/np.sum(~np.isnan(pentad)) * 100

    # Find the locations of the values less than 21 and plot them
    plt.imshow(pentad < 21)
    plt.colorbar()
    plt.title('Values less than 21')
    plt.show()

# Plot the fraction of values less than 6, 11, 21 for each pentad
plt.plot(fraction_less_than_6, label = 'Fraction < 6')
plt.plot(fracttion_less_than_11, label = 'Fraction < 11')
plt.plot(fraction_less_than_21, label = 'Fraction < 21')
xlabel = plt.xlabel('Pentad')
ylabel = plt.ylabel('Fraction (%)')
title = plt.title('Fraction of n_data values below threshold')
legend = plt.legend()
    
```
